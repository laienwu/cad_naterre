Récupérer dans une colonne Actor_Name le full_name des acteurs sous le format: first_name + " " + last_name


```python
con=engine.execute("select first_name, last_name, concat(first_name,' ',last_name) as Actor_name from actor;")
for row in con:
    print (row)
```

Récupérer dans une colonne Actor_Name le full_name des acteurs sous le format: first_name en minuscule + "." + last_name en majuscule

```python
con=engine.execute("select first_name, last_name, concat(lower(first_name),' ',upper(last_name)) as Actor_name from actor;")
for row in con:
    print (row)
```

Récupérer dans une colonne Actor_Name le full_name des acteurs sous le format: last_name en majuscule + "." + premiere lettre du first_name en majuscule

```python
con=engine.execute("select first_name, last_name, concat(concat(upper(left(first_name,1)),lower(substring(first_name,2))),' ' ,upper(last_name)) as Actor_name from actor;")
for row in con:
    print (row)
```

```python
con=engine.execute("SELECT CONCAT(UPPER(last_name), ' ',  CONCAT(LEFT(first_name,1),'',LOWER(SUBSTRING(first_name, 2, LENGTH(first_name) ) )) ) AS Actor_Name FROM actor;")
for row in con:
    print(row)
```

```python
con=engine.execute("SELECT CONCAT(last_name, ' ',ucase(left(first_name,1)),lcase(substring(first_name,2))) as Actor_Name FROM actor limit 5;")
for i in con:
    print (i)
```

```python
con=engine.execute('SELECT CONCAT (UPPER(last_name)," ", UPPER(LEFT(first_name,1)),LOWER(SUBSTRING(first_name, 2, CHAR_LENGTH(first_name))))FROM actor;')
for row in con:
    print (row)
```

Afficher les acteurs (actor_id, first_name, last_name, nbre char first_name, nbre char last_name )par ordre décroissant de longueur de prénoms

```python
con=engine.execute("select *, length(first_name) as nb_char_first_name,length(last_name) as nb_char_last_name from actor order by nb_char_first_name desc;")
for row in con:
    print (row)
```

Trouver le nombre des acteurs ayant le meme last_name
Afficher le resultat  par ordre décroissant

```python
con=engine.execute("select last_name, count(last_name) as count from actor group by last_name having count>1 order by last_name desc")
for i in con:
    print(i)
```

```python
con=engine.execute("select last_name, count(last_name) as cfn from actor group by last_name order by cfn desc;")
for row in con:
    print (row)
```

Trouver le nombre des acteurs ayant le meme last_name
Afficher UNIQUEMENT les last_names communs à au moins 3 acteurs 
Afficher par ordre alph. croissant


```python
con=engine.execute("select last_name, count(last_name) as cfn from actor group by last_name having(count(last_name)>=3)order by last_name asc;")
for row in con:
    print (row)
```

Trouver le nombre des acteurs ayant le meme first_name Afficher le resultat par ordre alph. croissant

```python
con=engine.execute("select first_name, count(first_name) as cfn from actor group by first_name order by first_name asc;")
for row in con:
    print (row)
```
